```python
import plotnine as gg
from plotnine.data import economics, huron, mpg
```

```python
economics.head()
```

```python
(gg.ggplot(economics) + gg.aes(x="date", y="pop") + gg.geom_line())
```

```python
gg.ggplot(mpg) + gg.aes(x="class", y="hwy") + gg.geom_point()
```

```python
gg.ggplot(mpg) + gg.aes(x="class", y="hwy") + gg.geom_jitter(width=0.2, height=0)
```

```python
(
    gg.ggplot(huron)
    + gg.aes(x="factor(decade)", y="level")
    + gg.geom_boxplot()
    + gg.labs(x="decade", y="water level")
)
```

```python
(
    gg.ggplot(mpg)
    + gg.aes(x="cyl", y="hwy")
    + gg.geom_point(gg.aes(color="class"))
    + gg.labs(x="engine cylinders", y="miles per gallon", color="vehicle class")
)
```

```python
myPlot = gg.ggplot(economics) + gg.aes(x="date", y="pop") + gg.geom_line()
myPlot.save("output/real-python_plotnine/myPlot.png", dpi=600)
```
