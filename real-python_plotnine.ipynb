```python
from plotnine.data import economics, mpg, huron
from plotnine import ggplot, aes, geom_line, geom_point, geom_jitter, geom_boxplot, labs
```

```python
economics.head()
```

```python
(ggplot(economics) + aes(x="date", y="pop") + geom_line())
```

```python
ggplot(mpg) + aes(x="class", y="hwy") + geom_point()
```

```python
ggplot(mpg) + aes(x="class", y="hwy") + geom_jitter(width=0.2, height=0)
```

```python
(
    ggplot(huron)
    + aes(x="factor(decade)", y="level")
    + geom_boxplot()
    + labs(x="decade", y="water level")
)
```

```python
(
    ggplot(mpg)
    + aes(x="cyl", y="hwy")
    + geom_point(aes(color="class"))
    + labs(x="engine cylinders", y="miles per gallon", color="vehicle class")
)
```

```python
myPlot = ggplot(economics) + aes(x="date", y="pop") + geom_line()
myPlot.save("output/real-python_plotnine/myPlot.png", dpi=600)
```
